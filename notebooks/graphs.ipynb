```python
import matplotlib.pyplot as plt
```

```python
import json
```

```python
x = list(range(0, 11))
```

```python
x
```

```python
with open("full_history.json", "r") as f:
    history = json.load(f)
```

```python
history
```

```python
plt.style.use("ggplot")

y1 = history["train_loss"]
y2 = history["val_loss"]

plt.plot(x, y1, label="Train Loss", )
plt.plot(x, y2, label = "Validation Loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.ylim(0, 12)
plt.xlim(0, 10)
plt.title('Graph of Validation and Training Loss during Ten Epochs\n')
plt.legend(loc="upper right", shadow = True, facecolor = "white", edgecolor = "black")
plt.grid(True, which='major', color='grey')
plt.grid(True, which='minor', color='#ADD8E6', alpha=0.5)
plt.minorticks_on()
plt.show()
```

```python
y1 = history["train_accuracy"]
y2 = history["val_accuracy"]

plt.plot(x, y1, label="Train Accuracy")
plt.plot(x, y2, label = "Test Accuracy")

plt.xlabel("Epochs")
plt.ylabel("Accuracy (%)")
plt.title('Train and Test Accuracy After 10 Epochs')
plt.legend(loc="lower right")
plt.show()
```
