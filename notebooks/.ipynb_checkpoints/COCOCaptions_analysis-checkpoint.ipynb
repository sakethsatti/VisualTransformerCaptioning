```python
import json
```

```python
with open("../../cocoannotations/captions_train2014.json", "r") as f: 
    annotations = json.load(f)
```

```python
annotations.keys()
```

```python
annotations["annotations"]
```

```python
file_names = []
captions = []

for dict_data in annotations["annotations"]:
    image_id = str(dict_data["image_id"])
    image_id = "COCO_train2014_" + ("0" * (12 - len(image_id))) + image_id + ".jpg"
    
    file_names.append(image_id)
    captions.append(dict_data["caption"])
```

```python
len(file_names)
```

```python
from PIL import Image
```

```python
Image.open("../../train2014/" + file_names[2000])
```

```python
import os
```

```python
captions[2000]
```

```python
count = 0
for x in file_names:
    if not (os.path.exists("../../train2014/" + x)):
        count += 1
```

```python
count
```

```python
import pandas as pd
```

```python
train_csv = pd.DataFrame({'path': file_names, 'captions': captions})
```

```python
train_csv.to_csv("cocotrain.csv", index = False)
```

```python
with open("../../cocoannotations/captions_val2014.json", "r") as f: 
    val_ann= json.load(f)
```
