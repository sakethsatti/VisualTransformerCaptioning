```python
def count_contractions(data, contractions):
  num_of_contractions = 0
  for cap in data.captions.values:
    for cont in contractions:
      if cont in cap:
        num_of_contractions += 1

  return num_of_contractions

print(count_contractions(data, contractions))
```

```python
def remove_contractions(data, contractions, expansion):
  for cap_idx in range(len(data.captions.values)):
    for cont_idx in range(len(contractions)):
      if contractions[cont_idx] in data.captions[cap_idx]:
        data.captions[cap_idx] = data.captions[cap_idx].replace(contractions[cont_idx], expansion[cont_idx])

  return data
```

```python
data = remove_contractions(data, contractions, expansion)
count_contractions(data, contractions)
```

```python
from tensorflow.keras import layers
```

```python
def get_vectorizer(tar):
    vectorizer = layers.TextVectorization(
        VOCAB_SIZE,
        output_sequence_length = 121,
        standardize=None,
    )
    vectorizer.adapt(list(tar))
    return vectorizer
```
